# 1. Lesson on how to search for Python commands

If you have a pandas DataFrame called df, then you can summarize its contents using df.info() or df.describe().  df.info() provides a list of column names with their counts and data types.  df.describe() will provide information such as the mean, min, max, standard deviation, and quantiles.  Thus:

```python
df = pd.DataFrame([[1, 4], [2, 5], [3, 6], [4, 7]], columns = ['A', 'B'])
df.describe()
```

In this describe() result, we see that the two columns A and B each have four elements.  The means and other statistics are shown.

8. What are the mean, median, and mode customer age in years?  (Rounding down to the next lower age.)
Are there any outliers?  (Customers with very large or very small ages, compared with the other ages?)

Suggested Google search or ChatGPT prompt: "how can I find out the mean, median, and mode of a pandas Series"

```python
# Create copy of deduplicated df
clean_df_bank = clean_df_bank.copy()

# Convert Birthdate field to age in years for each customer
today = pd.Timestamp('today')
clean_df_bank['Age'] = today.year - clean_df_bank['BirthDate'].dt.year - (
    (today.month < clean_df_bank['BirthDate'].dt.month) | 
    ((today.month == clean_df_bank['BirthDate'].dt.month) & (today.day < clean_df_bank['BirthDate'].dt.day))
)

# Specify Customer Age series for summary stats
customer_age = clean_df_bank['Age']

# Get mean, median, mode of customer age in years
print("Mean:", customer_age.mean())
print("Median:", customer_age.median())
print("Mode:", customer_age.mode().tolist()) 
```
